## Compare Functions

```python
from math import isclose


def ScienceForm(Num):
    return str(str("{:e}".format(Num))).replace("000e", "e").replace("0e", "e").replace("0e", "e").replace("e+00", "")



def Compare_Function(a, name_A, b, name_B):
    
    print("".join(["\nComparing ", color.UNDERLINE, "(", name_A, ")", color.END, " = ", color.BOLD, str(a), color.END, " to ", color.UNDERLINE, "(", name_B, ")", color.END, " = ", color.BOLD, str(b), color.END]))
    
    if(a == b):
        print("\tValues are identical...")
        return True

    abs_precision, rel_precision = 0, 0

    for digit in range(1, 11, 1):

        rel = 1/(10**digit)

        close_bool_abs = isclose(a, b, abs_tol=rel)
        # close_bool_rel = isclose(a, b, rel_tol=rel)
        # print("".join(["\nDigit accuracy = ", str(digit), " (Accurate up to ", str(rel*100), "%)"]))

        if(not close_bool_abs and abs_precision == 0):
            abs_precision = rel
            print("".join(["\tDigit Accuracy = ", color.BOLD, str(digit), " (Accurate up to ", str(rel), ")", color.END]))
            break

        # if(not close_bool_rel and rel_precision == 0):
        #     rel_precision = rel*100
        #     print("".join(["\tRelative (Percent) Accuracy = ", color.BOLD, str(rel*100), "%", color.END]))



        # if(not close_bool_abs and not close_bool_rel):
        #     # print("".join(["\nBoth 'False' after ", str(digit), " digits.\nMaximum Accuracy is either ", color.BOLD, ScienceForm(abs_precision), " digits",  color.END, " or ", color.BOLD, str(rel_precision), "% Difference", color.END]))
        #     break

    if(abs_precision == 0):
        print("".join([color.BOLD, "\n\tAbsolute Accuracy was not found... (Accuracy is better than ", str(rel), " digits)", color.END]))
    # if(rel_precision == 0):
    #     print("".join([color.BOLD, "\n\tRelative (Percent) Accuracy was not found... (Accuracy is better than ", str(rel*100), "% difference)", color.END]))
    
    print("")
        
        
print("\nCompare functions done.\n")
```

```python
def Monte_Carlo_Volume_Variance(Events_to_Gen, q2_xB_bin_list, N_loops, run_Q1, run_Q2, run_Q3, run_Q3_2):
    
    Hist_Full_List, Volume_Full_List, ave, std, error = {}, {}, {}, {}, {}

    # progress_tracker = 0
    
    for q2_xB_bin in q2_xB_bin_list:

        # print("".join(["Current Progress = ", str(round(progress_tracker/(len(q2_xB_bin_list)*N_loops), 4)), " % Done\n"]))

        for z_pT_bin in range(-1, z_pT_Border_Lines(q2_xB_bin)[2] + 1, 1):
            if((z_pT_bin == 0) or (q2_xB_bin != -1 and z_pT_bin == -1)):
                continue
                
            # if(z_pT_bin > 5 and z_pT_bin != -1):
            #     continue
                
            Actual_Volume = round(Bin_Volume_Calc("Histo_Main", q2_xB_bin, z_pT_bin), 6)
            Volume_Full_List[(q2_xB_bin, z_pT_bin)] = []
            
            if(run_Q3 == "yes"):
                max_vol = 1.75*Actual_Volume
                
                num_of_bins = 20
#                 num_of_bins = int(N_loops)# if(N_loops < 50) else int(N_loops*2)
                
#                 if(num_of_bins < 20):
#                     num_of_bins = 20
                    
#                 if(num_of_bins > 100):
#                     num_of_bins = 100
                    
                Hist_Full_List[(q2_xB_bin, z_pT_bin)] = ROOT.TH1F("".join(["test_histo_", str((q2_xB_bin, z_pT_bin))]), "".join(["Volume of Q^{2}-x_{B} Bin ", str(q2_xB_bin), " and z-p_{T} Bin ", str(z_pT_bin), "; Volume; Counts"]), num_of_bins, 0, max_vol)
            
            for loop_var in range(0, N_loops, 1):
                
                rdf_List = rdf_creation(Events_to_Gen)

                # if((z_pT_bin == 1) or (q2_xB_bin == -1 and z_pT_bin == -1)):
                #     progress_tracker += 1
                #     if(loop_var == 0):
                #         print("".join(["Current Progress = ", str(round(progress_tracker/(len(q2_xB_bin_list)*N_loops), 4)), " % Done\n"]))

                
                
                New_Volume = Monte_Carlo_Volume_Single(Events_to_Gen, rdf_List[0], rdf_List[1], q2_xB_bin, z_pT_bin)
                Volume_Full_List[(q2_xB_bin, z_pT_bin)].append(New_Volume)
                
                if(run_Q3 == "yes"):
                    Hist_Full_List[(q2_xB_bin, z_pT_bin)].Fill(New_Volume)
                    
                # if((loop_var == (N_loops - 1)) and (q2_xB_bin == q2_xB_bin_list[0]) and (z_pT_bin == -1)):
                #     # Runs these outputs in the last variance loop (but only the first bin loop)
                #     try:
                #         Final_Output_1(run_Q1, Events_to_Gen, rdf_List[0], rdf_List[1]).Draw()
                #     except:
                #         Final_Output_1(run_Q1, Events_to_Gen, rdf_List[0], rdf_List[1])
                #     try:
                #         Output_2 = Final_Output_2(run_Q2, Events_to_Gen, rdf_List[0], rdf_List[1], q2_xB_bin_list)
                #         for histo_2 in Output_2:
                #             histo_2.Draw()
                #     except:
                #         Final_Output_2(run_Q2, Events_to_Gen, rdf_List[0], rdf_List[1], q2_xB_bin_list)
            
            
            ave[(q2_xB_bin, z_pT_bin)] = round(stats.mean(Volume_Full_List[(q2_xB_bin, z_pT_bin)]), 6)
            std[(q2_xB_bin, z_pT_bin)] = round(stats.stdev(Volume_Full_List[(q2_xB_bin, z_pT_bin)]), 6)
            error[(q2_xB_bin, z_pT_bin)] = round(std[(q2_xB_bin, z_pT_bin)]/ROOT.sqrt(N_loops), 6)

            if(run_Q3_2 == "yes"):
                print("".join([color.BOLD, "The Average Calculated Volume of the Q^{2}-x_{B} Bin ", str(q2_xB_bin), " and z-p_{T} Bin ", str(z_pT_bin), " is: ", str(ave[(q2_xB_bin, z_pT_bin)]), "  ", str(error[(q2_xB_bin, z_pT_bin)]), color.END]))
                print("".join(["The Standard Deviation of the calculations is: ", str(std[(q2_xB_bin, z_pT_bin)])]))
                print("".join(["The Error of the calculations is:  ", str(error[(q2_xB_bin, z_pT_bin)])]))

                print("".join(["\nThe ACTUAL Volume of these bins is: ", str(Volume_Actual)]))

                if((Volume_Actual < (ave[(q2_xB_bin, z_pT_bin)] + error[(q2_xB_bin, z_pT_bin)])) and (Volume_Actual > (ave[(q2_xB_bin, z_pT_bin)] - error[(q2_xB_bin, z_pT_bin)]))):
                    print("The MC Calculation is within the correct range of the Actual Volume")
                else:
                    print("".join([color.BOLD, "Calculation is incorrect...", color.END]))

                # print("\n\n\n\n")
                print("\n\n")


    if(run_Q3 == "yes"):
        return [Hist_Full_List, Volume_Full_List, ave, std, error]
    else:
        return ["Hist_Full_List", Volume_Full_List, ave, std, error]

    
print("\n\nThis cell has finished running.\n\n")
```
