# DataFrame Creation Step(s)
See `cd_DataFramesSIDIS` for the Scripts

**Simple Instructions:**
 - Run `run_dataframe_makeROOT_helper.py` **in a tmux screen**
     - Will execute commands on `dataframe_makeROOT_epip_SIDIS.py`
     - Can check that main script's help message using `-hs`/`--help_script` in `run_dataframe_makeROOT_helper.py`
 - Choose which file set to run (i.e., `rdf`, `mdf`, or `gdf`) using the `-V`/`--variant` argument
     - Uses presets hard-coded at the top of `run_dataframe_makeROOT_helper.py` (edit as needed)
 - The `-j`/`--max_jobs` argument controls the number of parallel jobs the script will run at the same time
     - The run mode is set to `parallel` by default (is the best mode to use in pretty much all cases)
     - Can control this though with the `-m`/`--mode` argument
 - The `-mac`/`--matching_criteria` argument controls which matching criteria is used for the default kinematic calculations for the `mdf` files
     - Vary this to test the different criteria
 - The `-cn`/`--common_name` argument can be used to change the output file names
 - The `-o`/`--output_location` argument controls whether the output ROOT files go into the designated `work` or `volatile` directories
     - The paths are controlled by the `-w`/`-work`/`--work_dir` and `-vol`/`-volatile`/`--volatile_dir`
     - Use the default paths for best results (`work` for more important files, `volatile` for more temporary ones)
 - Log files produced while running go to a `scratch` directory controlled by `-sld`/`-scratch`/`--scratch_log_dir`
     - The `-plo`/`--primary_log_in_output` flag will cause the log files to be saved in the same location as the output ROOT files 
     - Use the default paths for best results (avoid writing the log files to the `volatile` folder)
 - Use `-e`/`--email` to recieve a notification when the jobs are all done running


**Approximate Run Times (based on observations — subject to change based on number of parallel jobs used)**
 - `rdf` --> ~30 minutes (with 15 parallel jobs — 171 total jobs)
 - `mdf` --> ~4 hours and 40 minutes (with 15 parallel jobs — 903 total jobs)
 - `gdf` --> ~1 hours and 40 minutes (with 30 parallel jobs — 903 total jobs)
