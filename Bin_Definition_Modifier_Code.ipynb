```python
%reset -f
import ROOT
# import numpy
import traceback

# import math
# import array
# import copy

from MyCommonAnalysisFunction_richcap import      Bin_Definition_Array, color
from Binning_Dictionaries             import Full_Bin_Definition_Array


def Compare_Arrays(bin_key, primary___array=Full_Bin_Definition_Array, secondary_array=Bin_Definition_Array):
    default_val_1 = primary___array.get(bin_key, None)
    default_val_2 = secondary_array.get(bin_key, None)
    if(default_val_1 == None):
        for ii in primary___array:
            if((bin_key in ii) and all(f"{bin_key}{jj}" != ii for jj in range(0, 10))):
                print(f"{color.Error}Needed to grab '{ii}' instead of '{bin_key}' from the 'primary___array'{color.END}")
                default_val_1 = primary___array[ii]
                break
    if(default_val_2 == None):
        for ii in secondary_array:
            if((bin_key in ii) and all(f"{bin_key}{jj}" != ii for jj in range(0, 10))):
                print(f"{color.Error}Needed to grab '{ii}' instead of '{bin_key}' from the 'secondary_array'{color.END}")
                default_val_2 = primary___array[ii]
                break
    return default_val_1, default_val_2
                

change_log = ""
def update_change_log(Q2y_Bin, zpTBin, log_input=change_log, z_max=None, z_min=None, pTmax=None, pTmin=None):
    log = f"For Q2-y Bin {Q2y_Bin} — z-pT Bin {zpTBin}, changed:\n"
    if(z_max != None):
        log = f"{log}\t(*) z_max to {z_max}\n"
    if(z_min != None):
        log = f"{log}\t(*) z_min to {z_min}\n"
    if(pTmax != None):
        log = f"{log}\t(*) pTmax to {pTmax}\n"
    if(pTmin != None):
        log = f"{log}\t(*) pTmin to {pTmin}\n"
    if("(*)" not in log):
        return log_input
    if(log not in log_input):
        log_input = f"{log_input}\n{log}"
    return log_input

def change_arrays(z_max_In, z_min_In, pTmax_In, pTmin_In, z_max_New=None, z_min_New=None, pTmax_New=None, pTmin_New=None):
    z_max = z_max_New if(z_max_New != None) else z_max_In
    z_min = z_min_New if(z_min_New != None) else z_min_In
    pTmax = pTmax_New if(pTmax_New != None) else pTmax_In
    pTmin = pTmin_New if(pTmin_New != None) else pTmin_In
    return [z_max, z_min, pTmax, pTmin]

print(f"\n{color.BBLUE}Ready to Run{color.END}\n")
```

```python
print(f"{color.BOLD}Total Num of Bins Defined in      {color.BBLUE}'Bin_Definition_Array'{color.END_B}: {color.UNDERLINE}{len(Bin_Definition_Array)}{color.END}")
print(f"{color.BOLD}Total Num of Bins Defined in {color.BBLUE}'Full_Bin_Definition_Array'{color.END_B}: {color.UNDERLINE}{len(Full_Bin_Definition_Array)}{color.END}\n\n")


# Q2_y_Bins = [4]
# z_pT_Bins = [1, 4]
# z_pT_Bins = list(range(1, 15))
Q2_y_Bins = [3]
z_pT_Bins = list(range(1, 7))
# z_pT_Bins = [14]


z_max = 0.68 # New for bin 1
z_max = 0.73 # New for bin 2, 3
z_max = None
z_min = None
pTmax = 0.98 # New for bin 1
pTmax = None
pTmin = None

for bin_key in Full_Bin_Definition_Array:
    if("Key" in bin_key):
        print(f"{color.BGREEN}{bin_key} --> {Full_Bin_Definition_Array[bin_key]}{color.END}")
        if("z-pT" in bin_key):
            print("")
    if(all(f"Q2-y={q2y}," not in bin_key for q2y in Q2_y_Bins) and (Q2_y_Bins != [])):
        continue
    if(("z-pT=" in bin_key) and any(q2y_only in z_pT_Bins for q2y_only in ["Q2_y_Only", "q2_y_only", "Q2y", "q2y"])):
        continue
    if((Q2_y_Bins != []) and (z_pT_Bins != [])):
        for q2y in Q2_y_Bins:
            skip = True
            if(f"Q2-y={q2y}" not in bin_key):
                skip = True
                continue
            for zpt in z_pT_Bins:
                if(f"Q2-y={q2y}, z-pT={zpt}" == bin_key):
                    skip = False
                    if(any(change != None for change in [z_max, z_min, pTmax, pTmin])):
                        val1, val2 = Compare_Arrays(bin_key)
                        z_max_1, z_min_1, pTmax_1, pTmin_1 = val1
                        z_max_2, z_min_2, pTmax_2, pTmin_2 = val2
                        Bin_Definition_Array[bin_key]      = change_arrays(z_max_1, z_min_1, pTmax_1, pTmin_1, z_max, z_min, pTmax, pTmin)
                        Full_Bin_Definition_Array[bin_key] = change_arrays(z_max_2, z_min_2, pTmax_2, pTmin_2, z_max, z_min, pTmax, pTmin)
                        change_log = update_change_log(q2y, zpt, change_log, z_max, z_min, pTmax, pTmin)
                    break
            if(not skip):
                break
        if(skip):
            continue
    val1, val2 = Compare_Arrays(bin_key, primary___array=Full_Bin_Definition_Array, secondary_array=Bin_Definition_Array)
    if(None not in [val1, val2]):
        print(f"\t{color.BBLUE}Bin_Definition_Array:      {color.END_B}{bin_key} -->{color.END} {val1}")
        print(f"\t{color.BBLUE}Full_Bin_Definition_Array: {color.END_B}{bin_key} -->{color.END} {val2}\n")
        
    
print(f"\n{color.BGREEN}Full List of Changes:{color.END}\n{change_log}\n")
```
